```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Define the file pattern 200_200J_100N_NFD_HN_NDJ_MPS_BW_TETRIS_FIFO
FILE_PATTERN = re.compile(
    # r'(\d+)_70J_50N_NFD_HN_NDJ_(SPS|MPS)_(BW|NBW)_(TETRIS|DRF|LIKELIHOOD|SGF|LGF|SEQ)_FIFO_topo\.csv'
    r'(\d+)_70J_50N_NFD_HN_NDJ_(SPS|MPS)_(BW|NBW)_(TETRIS)_FIFO_topo\.csv'
)

def compute_gini(array):
    """Compute the Gini index of a numpy array."""
    # Mean absolute difference
    mad = np.abs(np.subtract.outer(array, array)).mean()
    # Relative mean absolute difference
    mean = np.mean(array)
    if mean == 0:
        return np.nan
    rmad = mad / mean
    # Gini coefficient
    gini = 0.5 * rmad
    return gini

def compute_jain(array):
    """Compute Jain's fairness index of a numpy array."""
    numerator = np.sum(array) ** 2
    denominator = len(array) * np.sum(array ** 2)
    if denominator == 0:
        return np.nan
    return numerator / denominator

# === Define Specific Columns as per User's Request ===

# Spine Metrics
spine_cols_total = [
    'node_S0_total_bw', 'node_S1_total_bw'
]
spine_cols_reserved = [
    'node_S0_reserved_bw', 'node_S1_reserved_bw'
]
spine_cols_available = [
    'node_S0_available_bw', 'node_S1_available_bw'
]

# Leaf Metrics
leaf_cols_total = [
    'node_L0_total_bw', 'node_L1_total_bw', 'node_L2_total_bw', 
    'node_L3_total_bw', 'node_L4_total_bw'
]
leaf_cols_reserved = [
    'node_L0_reserved_bw', 'node_L1_reserved_bw', 'node_L2_reserved_bw', 
    'node_L3_reserved_bw', 'node_L4_reserved_bw'
]
leaf_cols_available = [
    'node_L0_available_bw', 'node_L1_available_bw', 'node_L2_available_bw', 
    'node_L3_available_bw', 'node_L4_available_bw'
]

# Node Links Utilization Metrics
node_links_cols_total = [
    # Edge L0
    'edge_L0-H0_total_bw', 'edge_L0-H1_total_bw', 'edge_L0-H2_total_bw', 
    'edge_L0-H3_total_bw', 'edge_L0-H4_total_bw', 'edge_L0-H5_total_bw', 
    'edge_L0-H6_total_bw', 'edge_L0-H7_total_bw', 'edge_L0-H8_total_bw', 
    'edge_L0-H9_total_bw',
    # Edge L1
    'edge_L1-H10_total_bw', 'edge_L1-H11_total_bw', 'edge_L1-H12_total_bw', 
    'edge_L1-H13_total_bw', 'edge_L1-H14_total_bw', 'edge_L1-H15_total_bw', 
    'edge_L1-H16_total_bw', 'edge_L1-H17_total_bw', 'edge_L1-H18_total_bw', 
    'edge_L1-H19_total_bw',
    # Edge L2
    'edge_L2-H20_total_bw', 'edge_L2-H21_total_bw', 'edge_L2-H22_total_bw', 
    'edge_L2-H23_total_bw', 'edge_L2-H24_total_bw', 'edge_L2-H25_total_bw', 
    'edge_L2-H26_total_bw', 'edge_L2-H27_total_bw', 'edge_L2-H28_total_bw', 
    'edge_L2-H29_total_bw',
    # Edge L3
    'edge_L3-H30_total_bw', 'edge_L3-H31_total_bw', 'edge_L3-H32_total_bw', 
    'edge_L3-H33_total_bw', 'edge_L3-H34_total_bw', 'edge_L3-H35_total_bw', 
    'edge_L3-H36_total_bw', 'edge_L3-H37_total_bw', 'edge_L3-H38_total_bw', 
    'edge_L3-H39_total_bw',
    # Edge L4
    'edge_L4-H40_total_bw', 'edge_L4-H41_total_bw', 'edge_L4-H42_total_bw', 
    'edge_L4-H43_total_bw', 'edge_L4-H44_total_bw', 'edge_L4-H45_total_bw', 
    'edge_L4-H46_total_bw', 'edge_L4-H47_total_bw', 'edge_L4-H48_total_bw', 
    'edge_L4-H49_total_bw'
]

node_links_cols_reserved = [
    # Edge L0
    'edge_L0-H0_reserved_bw', 'edge_L0-H1_reserved_bw', 'edge_L0-H2_reserved_bw', 
    'edge_L0-H3_reserved_bw', 'edge_L0-H4_reserved_bw', 'edge_L0-H5_reserved_bw', 
    'edge_L0-H6_reserved_bw', 'edge_L0-H7_reserved_bw', 'edge_L0-H8_reserved_bw', 
    'edge_L0-H9_reserved_bw',
    # Edge L1
    'edge_L1-H10_reserved_bw', 'edge_L1-H11_reserved_bw', 'edge_L1-H12_reserved_bw', 
    'edge_L1-H13_reserved_bw', 'edge_L1-H14_reserved_bw', 'edge_L1-H15_reserved_bw', 
    'edge_L1-H16_reserved_bw', 'edge_L1-H17_reserved_bw', 'edge_L1-H18_reserved_bw', 
    'edge_L1-H19_reserved_bw',
    # Edge L2
    'edge_L2-H20_reserved_bw', 'edge_L2-H21_reserved_bw', 'edge_L2-H22_reserved_bw', 
    'edge_L2-H23_reserved_bw', 'edge_L2-H24_reserved_bw', 'edge_L2-H25_reserved_bw', 
    'edge_L2-H26_reserved_bw', 'edge_L2-H27_reserved_bw', 'edge_L2-H28_reserved_bw', 
    'edge_L2-H29_reserved_bw',
    # Edge L3
    'edge_L3-H30_reserved_bw', 'edge_L3-H31_reserved_bw', 'edge_L3-H32_reserved_bw', 
    'edge_L3-H33_reserved_bw', 'edge_L3-H34_reserved_bw', 'edge_L3-H35_reserved_bw', 
    'edge_L3-H36_reserved_bw', 'edge_L3-H37_reserved_bw', 'edge_L3-H38_reserved_bw', 
    'edge_L3-H39_reserved_bw',
    # Edge L4
    'edge_L4-H40_reserved_bw', 'edge_L4-H41_reserved_bw', 'edge_L4-H42_reserved_bw', 
    'edge_L4-H43_reserved_bw', 'edge_L4-H44_reserved_bw', 'edge_L4-H45_reserved_bw', 
    'edge_L4-H46_reserved_bw', 'edge_L4-H47_reserved_bw', 'edge_L4-H48_reserved_bw', 
    'edge_L4-H49_reserved_bw'
]

node_links_cols_available = [
    # Edge L0
    'edge_L0-H0_available_bw', 'edge_L0-H1_available_bw', 'edge_L0-H2_available_bw', 
    'edge_L0-H3_available_bw', 'edge_L0-H4_available_bw', 'edge_L0-H5_available_bw', 
    'edge_L0-H6_available_bw', 'edge_L0-H7_available_bw', 'edge_L0-H8_available_bw', 
    'edge_L0-H9_available_bw',
    # Edge L1
    'edge_L1-H10_available_bw', 'edge_L1-H11_available_bw', 'edge_L1-H12_available_bw', 
    'edge_L1-H13_available_bw', 'edge_L1-H14_available_bw', 'edge_L1-H15_available_bw', 
    'edge_L1-H16_available_bw', 'edge_L1-H17_available_bw', 'edge_L1-H18_available_bw', 
    'edge_L1-H19_available_bw',
    # Edge L2
    'edge_L2-H20_available_bw', 'edge_L2-H21_available_bw', 'edge_L2-H22_available_bw', 
    'edge_L2-H23_available_bw', 'edge_L2-H24_available_bw', 'edge_L2-H25_available_bw', 
    'edge_L2-H26_available_bw', 'edge_L2-H27_available_bw', 'edge_L2-H28_available_bw', 
    'edge_L2-H29_available_bw',
    # Edge L3
    'edge_L3-H30_available_bw', 'edge_L3-H31_available_bw', 'edge_L3-H32_available_bw', 
    'edge_L3-H33_available_bw', 'edge_L3-H34_available_bw', 'edge_L3-H35_available_bw', 
    'edge_L3-H36_available_bw', 'edge_L3-H37_available_bw', 'edge_L3-H38_available_bw', 
    'edge_L3-H39_available_bw',
    # Edge L4
    'edge_L4-H40_available_bw', 'edge_L4-H41_available_bw', 'edge_L4-H42_available_bw', 
    'edge_L4-H43_available_bw', 'edge_L4-H44_available_bw', 'edge_L4-H45_available_bw', 
    'edge_L4-H46_available_bw', 'edge_L4-H47_available_bw', 'edge_L4-H48_available_bw', 
    'edge_L4-H49_available_bw'
]

# Combine Node Links Columns for Total and Reserved
node_links_cols = {
    'total_bw': node_links_cols_total,
    'reserved_bw': node_links_cols_reserved,
    'available_bw': node_links_cols_available
}

# Initialize data structure
data = {}

# List all files in the current directory
file_list = os.listdir('.')

for filename in file_list:
    match = FILE_PATTERN.match(filename)
    if match:
        print(match)
        id_str, sps_mps, bw_nbw, utility = match.groups()
        if id_str.startswith('433'):
            print(filename)
            # Process this file
            key = sps_mps  # Group by SPS or MPS
            if key not in data:
                data[key] = {}
            if utility not in data[key]:
                data[key][utility] = []
            # Load the CSV
            df = pd.read_csv(filename)
            # Divide relevant columns by 100 to get Gbps
            # Exclude 'allocation_step' and 'job_id'
            cols_to_divide = df.columns.difference(['allocation_step', 'job_id'])
            df[cols_to_divide] = df[cols_to_divide] / 100
            
            # === Compute Total Reserved Bandwidth and Total Bandwidth ===
            # Using the explicitly defined columns
            # Total Reserved Bandwidth
            df['total_reserved_bw'] = df[spine_cols_reserved + leaf_cols_reserved + node_links_cols['reserved_bw']].sum(axis=1)
            # Total Bandwidth
            df['total_bw'] = df[spine_cols_total + leaf_cols_total + node_links_cols['total_bw']].sum(axis=1)
            # Utilization
            df['utilization'] = df['total_reserved_bw'] / df['total_bw']
            
            # === Compute Per-Link Utilization ===
            # Initialize list to store per-link utilization column names
            per_link_util_cols = []
            
            # Spine Links Utilization
            for reserved, total in zip(spine_cols_reserved, spine_cols_total):
                util_col = reserved.replace('_reserved_bw', '_utilization')
                df[util_col] = df[reserved] / df[total]
                per_link_util_cols.append(util_col)
            
            # Leaf Links Utilization
            for reserved, total in zip(leaf_cols_reserved, leaf_cols_total):
                util_col = reserved.replace('_reserved_bw', '_utilization')
                df[util_col] = df[reserved] / df[total]
                per_link_util_cols.append(util_col)
            
            # Node Links Utilization
            for reserved, total in zip(node_links_cols_reserved, node_links_cols_total):
                util_col = reserved.replace('_reserved_bw', '_utilization')
                df[util_col] = df[reserved] / df[total]
                per_link_util_cols.append(util_col)
            
            # Replace NaN and inf with 0
            df[per_link_util_cols] = df[per_link_util_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
            
            # === Compute Fairness Metrics ===
            # Fairness is computed across per-link utilizations
            fairness_util_cols = per_link_util_cols  # Only the specific per-link utilizations
            
            def compute_fairness_metrics(row):
                utilizations = row[fairness_util_cols].values
                utilizations = utilizations[~np.isnan(utilizations)]
                utilizations = utilizations[utilizations >= 0]
                if len(utilizations) == 0:
                    return pd.Series({'gini': np.nan, 'jain': np.nan})
                gini = compute_gini(utilizations)
                jain = compute_jain(utilizations)
                return pd.Series({'gini': gini, 'jain': jain})
            
            fairness_metrics = df.apply(compute_fairness_metrics, axis=1)
            df['gini'] = fairness_metrics['gini']
            df['jain'] = fairness_metrics['jain']
            
            # === Compute Specific Utilizations ===
            # Spine Switch Utilization
            df['spine_reserved_bw'] = df[spine_cols_reserved].sum(axis=1)
            df['spine_total_bw'] = df[spine_cols_total].sum(axis=1)
            df['spine_utilization'] = df['spine_reserved_bw'] / df['spine_total_bw']
            
            # Leaf Switch Utilization
            df['leaf_reserved_bw'] = df[leaf_cols_reserved].sum(axis=1)
            df['leaf_total_bw'] = df[leaf_cols_total].sum(axis=1)
            df['leaf_utilization'] = df['leaf_reserved_bw'] / df['leaf_total_bw']
            
            # Node Links Utilization (Already computed per-link; here, compute aggregate if needed)
            df['node_reserved_bw'] = df[node_links_cols['reserved_bw']].sum(axis=1)
            df['node_total_bw'] = df[node_links_cols['total_bw']].sum(axis=1)
            df['node_utilization'] = df['node_reserved_bw'] / df['node_total_bw']
            
            # Store the DataFrame
            data[key][utility].append(df)

# Plotting
for sps_mps in data:
    # Overall Utilization
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        # Concatenate DataFrames for the utility
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        # Group by allocation_step and compute mean utilization
        df_grouped = df.groupby('allocation_step')['utilization'].mean().reset_index()
        # Smooth the curve using rolling mean
        df_grouped['utilization_smooth'] = df_grouped['utilization'].rolling(window=5, min_periods=1).mean()
        # Plot
        plt.plot(df_grouped['allocation_step'], df_grouped['utilization_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel('Utilization')
    plt.title(f'Average Utilization over Time - {sps_mps}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Gini Index
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        # Group by allocation_step and compute mean Gini index
        df_grouped = df.groupby('allocation_step')['gini'].mean().reset_index()
        # Smooth the curve
        df_grouped['gini_smooth'] = df_grouped['gini'].rolling(window=5, min_periods=1).mean()
        plt.plot(df_grouped['allocation_step'], df_grouped['gini_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel('Gini Index')
    plt.title(f'Gini Index over Time - {sps_mps}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Jain's Fairness Index
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        # Group by allocation_step and compute mean Jain's index
        df_grouped = df.groupby('allocation_step')['jain'].mean().reset_index()
        # Smooth the curve
        df_grouped['jain_smooth'] = df_grouped['jain'].rolling(window=5, min_periods=1).mean()
        plt.plot(df_grouped['allocation_step'], df_grouped['jain_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel("Jain's Fairness Index")
    plt.title(f"Jain's Fairness Index over Time - {sps_mps}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Spine Switch Utilization
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        df_grouped = df.groupby('allocation_step')['spine_utilization'].mean().reset_index()
        df_grouped['utilization_smooth'] = df_grouped['spine_utilization'].rolling(window=5, min_periods=1).mean()
        plt.plot(df_grouped['allocation_step'], df_grouped['utilization_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel('Spine Switch Utilization')
    plt.title(f'Spine Switch Utilization over Time - {sps_mps}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Leaf Switch Utilization
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        df_grouped = df.groupby('allocation_step')['leaf_utilization'].mean().reset_index()
        df_grouped['utilization_smooth'] = df_grouped['leaf_utilization'].rolling(window=5, min_periods=1).mean()
        plt.plot(df_grouped['allocation_step'], df_grouped['utilization_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel('Leaf Switch Utilization')
    plt.title(f'Leaf Switch Utilization over Time - {sps_mps}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Node Links Utilization
    plt.figure(figsize=(10, 6))
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        df_grouped = df.groupby('allocation_step')['node_utilization'].mean().reset_index()
        df_grouped['utilization_smooth'] = df_grouped['node_utilization'].rolling(window=5, min_periods=1).mean()
        plt.plot(df_grouped['allocation_step'], df_grouped['utilization_smooth'], label=utility)
    plt.xlabel('Allocation Step')
    plt.ylabel('Node Links Utilization')
    plt.title(f'Node Links Utilization over Time - {sps_mps}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # === New Plot: Each Single Link Utilization Over Time ===
    plt.figure(figsize=(20, 12))  # Increased size to accommodate many lines
    for utility in data[sps_mps]:
        df_list = data[sps_mps][utility]
        df = pd.concat(df_list, ignore_index=True)
        # Identify all per-link utilization columns (specific to user)
        # These are the 'utilization' columns we computed earlier
        per_link_util_cols = [
            col for col in df.columns 
            if '_utilization' in col and 
            col not in ['utilization', 'spine_utilization', 'leaf_utilization', 'node_utilization']
        ]
        # Group by allocation_step and compute mean utilization for each link
        df_grouped = df.groupby('allocation_step')[per_link_util_cols].mean().reset_index()
        # Smooth the utilization curves
        df_grouped[per_link_util_cols] = df_grouped[per_link_util_cols].rolling(window=5, min_periods=1).mean()
        # Plot each link's utilization
        for col in per_link_util_cols:
            plt.plot(df_grouped['allocation_step'], df_grouped[col], label=f"{utility} - {col}")
    
    plt.xlabel('Allocation Step')
    plt.ylabel('Link Utilization')
    plt.title(f'Per-Link Utilization over Time - {sps_mps}')
    # Due to a large number of lines, adjust the legend accordingly
    # To prevent the legend from being too cluttered, we can limit the number of entries or make it interactive
    # Here, we'll place it outside the plot
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small', ncol=1)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import os
from pathlib import Path

# Define the regex pattern
FILE_PATTERN = re.compile(
    r'^(\d+)_70J_50N_NFD_HN_NDJ_(SPS|MPS)_(BW|NBW)_(TETRIS|DRF|LIKELIHOOD|SGF|LGF|SEQ)_FIFO\.csv$'
)

def find_matching_files(directory, starting_number, prefix_match=False):
    matching_files = []
    p = Path(directory)
    for file in p.iterdir():
        if file.is_file():
            match = FILE_PATTERN.match(file.name)
            if match:
                print(f"Matched filename: {file.name}")
                number = match.group(1)
                if (number == starting_number) if not prefix_match else number.startswith(starting_number):
                    sps_mps = match.group(2)
                    utility_function = match.group(4)
                    matching_files.append({
                        'filename': file.name,
                        'sps_mps': sps_mps,
                        'utility_function': utility_function
                    })
                    print(f"Added to matching files: {file.name}")
    return matching_files

def process_files(directory, starting_number, prefix_match=False):
    matching_files = find_matching_files(directory, starting_number, prefix_match)
    
    if matching_files:
        print("\nMatching files:")
        for file in matching_files:
            print(file)
    else:
        print("\nNo matching files found.")
    
    print(matching_files)
    
    # Initialize a dictionary to store utilization data
    utilization_data = {}
    
    for exp in matching_files:
        filename = exp['filename']
        sps_mps = exp['sps_mps']
        utility_function = exp['utility_function']
        print(f"\nProcessing file: {filename}")
        
        try:
            df = pd.read_csv(Path(directory) / filename)
        except FileNotFoundError:
            print(f"File not found: {filename}")
            continue
        except pd.errors.EmptyDataError:
            print(f"No data: {filename}")
            continue
        except Exception as e:
            print(f"Error reading {filename}: {e}")
            continue
        
        # Validate required columns
        required_columns = ['time_instant']
        if not all(col in df.columns for col in required_columns):
            print(f"Missing required columns in {filename}")
            continue
        
        # Extract time_instant
        time_instant = df['time_instant']
        
        # Get GPU initial and used columns
        gpu_initial_columns = [col for col in df.columns if '_initial_gpu' in col]
        gpu_used_columns = [col for col in df.columns if '_used_gpu' in col]
        
        # Compute total initial and used GPU
        total_gpu_initial = df[gpu_initial_columns].sum(axis=1)
        total_gpu_used = df[gpu_used_columns].sum(axis=1)
        
        # Compute GPU utilization percentage
        gpu_utilization = total_gpu_used / total_gpu_initial * 100
        
        # Similarly for CPU
        cpu_initial_columns = [col for col in df.columns if '_initial_cpu' in col]
        cpu_used_columns = [col for col in df.columns if '_used_cpu' in col]
        total_cpu_initial = df[cpu_initial_columns].sum(axis=1)
        total_cpu_used = df[cpu_used_columns].sum(axis=1)
        cpu_utilization = total_cpu_used / total_cpu_initial * 100
        
        # Similarly for BW
        bw_initial_columns = [col for col in df.columns if '_initial_bw' in col]
        bw_used_columns = [col for col in df.columns if '_used_bw' in col]
        total_bw_initial = df[bw_initial_columns].sum(axis=1)
        total_bw_used = df[bw_used_columns].sum(axis=1)
        
        # Handle zero initial bandwidth to avoid division by zero
        total_bw_initial = total_bw_initial.replace(0, np.nan)
        bw_utilization = total_bw_used / total_bw_initial * 100
        
        # Store the utilization data
        if sps_mps not in utilization_data:
            utilization_data[sps_mps] = {}
        if utility_function not in utilization_data[sps_mps]:
            utilization_data[sps_mps][utility_function] = {}
        
        utilization_data[sps_mps][utility_function]['time_instant'] = time_instant
        utilization_data[sps_mps][utility_function]['gpu_utilization'] = gpu_utilization
        utilization_data[sps_mps][utility_function]['cpu_utilization'] = cpu_utilization
        utilization_data[sps_mps][utility_function]['bw_utilization'] = bw_utilization
        utilization_data[sps_mps][utility_function]['total_gpu_initial'] = total_gpu_initial
        utilization_data[sps_mps][utility_function]['total_gpu_used'] = total_gpu_used
        utilization_data[sps_mps][utility_function]['total_cpu_initial'] = total_cpu_initial
        utilization_data[sps_mps][utility_function]['total_cpu_used'] = total_cpu_used
        utilization_data[sps_mps][utility_function]['total_bw_initial'] = total_bw_initial
        utilization_data[sps_mps][utility_function]['total_bw_used'] = total_bw_used
    
    # Now, generate plots
    for sps_mps in ['SPS', 'MPS']:
        if sps_mps not in utilization_data:
            continue
        for resource in ['gpu', 'cpu', 'bw']:
            plt.figure(figsize=(10, 6))
            for utility_function in utilization_data[sps_mps]:
                time_instant = utilization_data[sps_mps][utility_function]['time_instant']
                utilization = utilization_data[sps_mps][utility_function][f'{resource}_utilization']
                plt.plot(time_instant, utilization, label=utility_function)
            plt.title(f'{sps_mps} - {resource.upper()} Utilization Over Time')
            plt.xlabel('Time Instant')
            plt.ylabel('Utilization (%)')
            plt.legend()
            plt.grid(True)
            plt.savefig(f'{sps_mps}_{resource}_utilization.png')
            plt.show()
    
            # Plot total used resources over time
            plt.figure(figsize=(10, 6))
            for utility_function in utilization_data[sps_mps]:
                time_instant = utilization_data[sps_mps][utility_function]['time_instant']
                total_used = utilization_data[sps_mps][utility_function][f'total_{resource}_used']
                plt.plot(time_instant, total_used, label=utility_function)
            plt.title(f'{sps_mps} - Total {resource.upper()} Used Over Time')
            plt.xlabel('Time Instant')
            plt.ylabel(f'Total {resource.upper()} Used')
            plt.legend()
            plt.grid(True)
            plt.savefig(f'{sps_mps}_{resource}_used.png')
            plt.show()

# Example usage
if __name__ == "__main__":
    directory = '.'  # Current directory
    starting_number = '1'  # Pass as string
    prefix_match = False  # Set to True if you want to match files starting with '1', '10', '11', etc.
    
    process_files(directory, starting_number, prefix_match)
```
